```python
#top customer by revenue
df.groupby(['Category','Region'])['Revenue'].sum().sort_values(ascending=False)
```

```python
#Yearly Revenue
yearly_Revenue=df.groupby(['Year'])['Revenue'].sum().sort_values(ascending=False)
```

```python
yearly_Revenue
```

```python
#which category generate most revenue
df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
#Most profitable category
df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
#profit margin
df['profit_margin']=df['Profit']/df['Revenue']*100
```

```python
#Profit margin by category
df.groupby('Category')['profit_margin'].mean().sort_values(ascending=False).map("{:.1f}%".format)
```

```python
df=df.reset_index(drop=True)
df.head(10)
```

```python
#Best region by revenue
df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
```

```python
#Best region by profit
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
#Which category has high ad spend & But low profit
df.groupby('Category')[['Profit' , 'Ad_Spend']].sum().sort_values(by='Profit',ascending=False)
```

```python
#AD_spend increase Revenue
df[['Ad_Spend','Revenue']].corr()
```

```python
#Create Roas column
#Roas=For that how much return ads generate
df['Roas']=df['Revenue']/df['Ad_Spend']
```

```python
df.head()
```

```python
df.groupby('Category')['Roas'].mean().sort_values(ascending=False)
```

```python
#which category has  avg Rating.
df.groupby('Category')['Customer_Rating'].mean().sort_values(ascending=False)
```

```python
#Rating impact on Revenue
df[['Customer_Rating', 'Revenue']].corr()
```
